# Example usage

Here we will demonstrate how to use `pycounts` to count the words in a text dile and plot the top 5 results.

## Imports

```python
from pycounts_wt.pycounts_wt import count_words
from pycounts_wt.plotting import plot_words
```

## Create a text file

We'll first create a text file to work with using a famous quote from Einstein:

```python
quote = """Insanity is doing the same thing
over and over and expecting different results."""
with open("einstein.txt", "w") as file:
    file.write(quote)
```

## Count words

we can count the words in our text file using the `count_words()` function. Note that the function removes punctuations and makes all words lowercase before Counting.

```python
counts = count_words("einstein.txt")
print(counts)
```

## Plot words

We can now plot the result using the `plot_words()` function:

```python
fig = plot_words(counts, n=5)
```
